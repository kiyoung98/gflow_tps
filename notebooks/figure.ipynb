```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.tri as mtri

def energy_landscape(x, y):
    # Define basin 1 (centered at (1, 0)) with a steeper slope
    basin1 = -12 * np.exp(-((x - 1)**2 + (y - 1)**2) / 1)

    # Define basin 2 (centered at (-1, 0)) with a wider and shallower minimum
    basin2 = -9 * np.exp(-((x + 0.3)**2 + (y + 1)**2) / 1)
    
    basin3 = -11 * np.exp(-((x + 1.2)**2 + (y-1)**2) / 1)

    # Combine the effects, with basin 1 having a slightly higher energy barrier
    return basin1 + basin2 + basin3

fig = plt.figure(figsize=plt.figaspect(0.5))

# Meshgrid for parameterization
u = np.linspace(-2, 2, endpoint=True, num=50)
v = np.linspace(-2, 2, endpoint=True, num=50)
u, v = np.meshgrid(u, v)

z = energy_landscape(u, v)

tri = mtri.Triangulation(u.flatten(), v.flatten())

ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_trisurf(u.flatten(), v.flatten(), z.flatten(), triangles=tri.triangles, cmap=plt.cm.plasma)  

ax.set_axis_off()
ax.view_init(elev=70, azim=120, roll=10)

plt.tight_layout()
plt.show()
```
